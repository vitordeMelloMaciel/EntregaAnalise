```python
# ===================== An√°lise Explorat√≥ria - Banco do Primeiro =====================
# Este notebook conecta ao banco do primeiro (que ainda existe) e gera EDA organizado para cada tabela.

# --------------------- Importa√ß√µes ---------------------
import psycopg2
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Configura√ß√µes de visualiza√ß√£o
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (8,5)

# --------------------- URL do banco direto ---------------------
DATABASE_URL = "postgresql://neondb_owner:npg_StuVDjnZ8L3v@ep-still-cake-ad03wttb.c-2.us-east-1.aws.neon.tech/frotaviva?sslmode=require&channel_binding=require"

# --------------------- Conex√£o com o banco ---------------------
conn = psycopg2.connect(DATABASE_URL, sslmode="require")

# --------------------- Obt√©m todas as tabelas ---------------------
query_tables = """
SELECT table_name
FROM information_schema.tables
WHERE table_schema = 'public'
AND table_type = 'BASE TABLE';
"""

tables = pd.read_sql(query_tables, conn)['table_name'].tolist()
print("Tabelas encontradas:", tables)

# --------------------- Fun√ß√£o para EDA ---------------------
def eda_tabela(tabela, conn, limite_linhas=5000):
    print(f"\n\n===== TABELA: {tabela} =====")
    
    # L√™ a tabela com limite de linhas (evita travamentos)
    df = pd.read_sql(f"SELECT * FROM {tabela} LIMIT {limite_linhas};", conn)
    
    # Verifica se est√° vazia
    if df.empty:
        print(" Tabela vazia, pulando...")
        return None

    # --------------------- Ignora colunas in√∫teis ---------------------
    colunas_inuteis = [
        'id', 'id_', 'idcliente', 'id_cliente', 'id_motorista', 'id_caminhao',
        'cliente_id', 'pedido_id', 'codigo', 'codigo_produto', 'codigo_empresa',
        'telefone', 'email', 'cpf', 'cnpj', 'descricao', 'observacao',
        'comentario', 'created_at', 'updated_at', 'data_criacao', 'data_atualizacao'
    ]
    removidas = [c for c in df.columns if c.lower() in colunas_inuteis]
    df = df.drop(columns=removidas, errors='ignore')

    # Remove colunas sem varia√ß√£o
    df = df.loc[:, df.nunique() > 1]

    if removidas:
        print(f"üîπ Colunas ignoradas: {removidas}")

    # Se a tabela ficar sem colunas, pula a an√°lise
    if df.shape[1] == 0:
        print(" Nenhuma coluna relevante restante ap√≥s remo√ß√£o de colunas in√∫teis. Pulando an√°lise.\n")
        return None
    # ------------------------------------------------------------------------------

    # Informa√ß√µes gerais
    print("\n--- Informa√ß√µes gerais ---")
    print(df.info())
    
    # Estat√≠sticas num√©ricas
    print("\n--- Estat√≠sticas num√©ricas ---")
    print(df.describe())
    
    # Estat√≠sticas categ√≥ricas (s√≥ se houver colunas object)
    cat_cols = df.select_dtypes(include=['object']).columns
    print("\n--- Estat√≠sticas categ√≥ricas ---")
    if len(cat_cols) > 0:
        print(df[cat_cols].describe())
    else:
        print("Nenhuma coluna categ√≥rica encontrada.")
    
    # Valores ausentes
    print("\n--- Valores ausentes ---")
    print(df.isnull().sum())
    
    # --------------------- Gr√°ficos ---------------------
    num_cols = df.select_dtypes(include=['int64', 'float64']).columns
    
    # Distribui√ß√£o num√©rica
    for col in num_cols:
        plt.figure()
        sns.histplot(df[col], kde=True, bins=20)
        plt.title(f'Distribui√ß√£o de {col} ({tabela})')
        plt.show()
    
    # Contagem de categorias (s√≥ se tiver poucas categorias)
    for col in cat_cols:
        if df[col].nunique() < 30:
            plt.figure()
            sns.countplot(y=col, data=df)
            plt.title(f'Contagem de categorias em {col} ({tabela})')
            plt.show()
    
    # Matriz de correla√ß√£o
    if len(num_cols) > 1:
        plt.figure(figsize=(10,8))
        sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm')
        plt.title(f'Matriz de correla√ß√£o ({tabela})')
        plt.show()
    
    # Exporta para CSV (opcional)
    df.to_csv(f"{tabela}_EDA.csv", index=False)
    
    return df

# --------------------- Executa EDA para todas as tabelas ---------------------
dfs = {}
for tabela in tables:
    print(f"\n‚û°Ô∏è Analisando tabela: {tabela}")
    dfs[tabela] = eda_tabela(tabela, conn)

# --------------------- Fecha a conex√£o ---------------------
conn.close()

print("\n‚úÖ An√°lise conclu√≠da! Todos os DataFrames est√£o salvos no dicion√°rio 'dfs' e exportados como CSV.")
```
